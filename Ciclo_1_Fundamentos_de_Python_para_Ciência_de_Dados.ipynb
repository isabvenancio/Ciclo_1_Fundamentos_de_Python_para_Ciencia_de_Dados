## 2. Calculo de estatística com Numpy


### 2.1 Idade Média dos Pacientes

```python
idades = df['age'].to_numpy()
idade_media = np.mean(idades)
idade_desvio = np.std(idades)
idade_cv = idade_desvio / idade_media
idade_mediana = np.median(idades)
idade_min = np.min(idades)
idade_max = np.max(idades)
print(f"Idade Média: {idade_media:} anos")
print(f"Desvio Padrão: {idade_desvio:} anos")
print(f"CV: {idade_cv * 100:.2f}%")
print(f"Mediana: {idade_mediana:} anos")
print(f"Idade mínima: {idade_min:} anos")
print(f"Idade máxima: {idade_max:} anos")
```

### 2.2 Estatísticas da variável `oldpeak`

```python
oldpeak = df['oldpeak'].to_numpy()
oldpeak_medio = np.mean(oldpeak)
oldpeak_desvio = np.std(oldpeak)
oldpeak_mediana = np.median(oldpeak)
oldpeak_min = np.min(oldpeak)
oldpeak_max = np.max(oldpeak)

print('\nMédia de oldpeak:', oldpeak_medio.round(4))
print('Desvio padrão de oldpeak:', oldpeak_desvio.round(4))
print('Mediana de oldpeak:', oldpeak_mediana)
print('Oldpeak mínimo:', oldpeak_min)
print('Oldpeak máximo:', oldpeak_max)
```

## 4. Filtragem e groupby para comparação de médias de variáveis entre pacientes com e sem doença cardíaca

### 4.1 Idade média dos pacientes com e sem doença cardíaca

```python
media_idade_groupby = df.groupby('target')['age'].mean()
display(media_idade_groupby)

idades_doentes = df.loc[df.target == 1, "age"]
media_doentes = np.mean(idades_doentes)
desvio_doentes = np.std(idades_doentes)
cv_doentes = desvio_doentes / media_doentes

idades_saudaveis = df.loc[df.target == 0, "age"]
media_saudaveis = np.mean(idades_saudaveis)
desvio_saudaveis = np.std(idades_saudaveis)
cv_saudaveis = desvio_saudaveis / media_saudaveis

print("\nCOM DOENÇA CARDÍACA")
print(f"Idade Média: {media_doentes:} anos")
print(f"Desvio Padrão: {desvio_doentes:} anos")
print(f"CV: {cv_doentes * 100:.2f}%")

print("\nSEM DOENÇA CARDÍACA")
print(f"Idade Média: {media_saudaveis:} anos")
print(f"Desvio Padrão: {desvio_saudaveis:} anos")
print(f"CV: {cv_saudaveis * 100:.2f}%")
```

### 4.2 Oldpeak médio dos pacientes com e sem doença cardíaca

```python
df_oldpeak_target = df.groupby('target', observed=True)['oldpeak'].mean().round(4)

display(df_oldpeak_target)

op_sem = df_oldpeak_target.loc[0]
op_com = df_oldpeak_target.loc[1]
relacao_op = op_com / op_sem

oldpeak_sem = df[df['target'] == 0]['oldpeak']
oldpeak_com = df[df['target'] == 1]['oldpeak']

q1 = df['oldpeak'].quantile(0.25)
q2 = df['oldpeak'].quantile(0.5)
q3 = df['oldpeak'].quantile(0.75)
intervalo = q3 - q1
ls = q3 + 1.5 * intervalo
li = q1 - 1.5 * intervalo

q1_sem = oldpeak_sem.quantile(0.25)
q2_sem = oldpeak_sem.quantile(0.5)
q3_sem = oldpeak_sem.quantile(0.75)
intervalo_sem = q3_sem - q1_sem
ls_sem = q3_sem + 1.5 * intervalo_sem
li_sem = q1_sem - 1.5 * intervalo_sem

q1_com = oldpeak_com.quantile(0.25)
q2_com = oldpeak_com.quantile(0.5)
q3_com = oldpeak_com.quantile(0.75)
intervalo_com = q3_com - q1_com
ls_com = q3_com + 1.5 * intervalo_com
li_com = q1_com - 1.5 * intervalo_com

outliers_sem = oldpeak_sem[(oldpeak_sem > ls_sem) | (oldpeak_sem < li_sem)].count()

outliers_com = oldpeak_com[(oldpeak_com > ls_com) | (oldpeak_com < li_com)].count()

print('\nOldpeak médio do paciente sem doença cardíaca:', op_sem)
print('Oldpeak médio do paciente com doença cardíaca:', op_com)
print(f'Pacientes com doenças cardíacas apresentam um oldpeak {relacao_op.round(4)} vezes maior que pacientes sem a doença\n')

print(f'Quartis 1, 2 e 3 dos oldpeaks gerais: {q1}, {q2}, {q3}')
print(f'Limites dos oldpeaks gerais: {li.round(4)} a {ls}\n')

print(f'Quartis 1, 2 e 3 dos oldpeaks sem doença cardíaca: {q1_sem}, {q2_sem}, {q3_sem}')
print(f'Limites dos oldpeaks sem doença cardíaca: {li_sem.round(4)} a {ls_sem.round(4)}')
print('Número de outliers dos oldpeaks sem doença cardíaca:', outliers_sem,'\n')

print(f'Quartis 1, 2 e 3 dos oldpeaks com doença cardíaca: {q1_com}, {q2_com}, {q3_com}')
print(f'Limites dos oldpeaks com doença cardíaca: {li_com.round(4)} a {ls_com}')
print('Número de outliers dos oldpeaks com doença cardíaca:', outliers_com)
```

### 5.3 Gráfico de barras comparando a prevalência da doença por sexo

### 5.4 Gráfico de barras mostrando a prevalência da doença por faixa etária e sexo

Criação das faixas etárias

```python
menor_idade = df.age.min()
maior_idade = df.age.max()

df["faixa"] = pd.cut(
    df["age"],
    bins= np.linspace(menor_idade, maior_idade, 9),
    include_lowest=True
) #inserção de uma nova coluna, indicando a faixa etária do paciente

doentes = df.loc[df.target == 1, ['sex','faixa']]

homens = doentes.loc[doentes.sex == 1, 'faixa'].value_counts(sort=False) #poderia ser feito um sort_index, mas os intervalos já estão ordenados, então basta não alterar a ordem
mulheres = doentes.loc[doentes.sex == 0, 'faixa'].value_counts(sort=False)

doentes_totais = len(df[df.target==1])
homens *= 100/doentes_totais
mulheres *= 100/doentes_totais
```

Interpretação: O histograma revela que a prevalência da doença cresce com a idade, concentrando-se nas faixas de 53-59 e 59-65 anos, especialmente entre homens. As faixas mais jovens (28-47 anos) e a mais idosa (71-77 anos) apresentam baixa representatividade. Isso indica associação entre a doença e o envelhecimento, com possível maior predisposição masculina na meia-idade avançada.

```python
import matplotlib.pyplot as plt

x = homens.index.map(lambda inter: f"{int(inter.left)}-{int(inter.right)}")

plt.figure(figsize=(12,6))
plt.bar(x,homens, label= "Homens")
plt.bar(x, mulheres, bottom = homens, color= "red", label = "Mulheres")
plt.legend()

plt.xticks(rotation=30)

plt.title("Distribuição de doentes por idade e sexo", fontsize= 20)
plt.xlabel("Faixa etária",  fontsize = 15)
plt.ylabel("Representatividade(em %)", fontsize = 15)

plt.yticks(ticks= range(0,101, 10))
plt.grid(axis="y",linestyle= "--")
plt.axvspan(2.6,5.4,alpha=0.1, color= 'green')

plt.text(2.7, 55, "Maior concentração de casos\n", color= "blue", fontsize= 14)

plt.box(False)
plt.tick_params(length=0)

plt.show()
```
